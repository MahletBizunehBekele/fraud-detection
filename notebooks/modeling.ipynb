```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
import pandas as pd

df = pd.read_csv(
    "../data/processed/fraud_processed.csv"
)
```

```python
X = df.drop(
    columns=[
        'class',
        'signup_time',
        'purchase_time'
    ]
)

y = df['class']
```

```python
X = X.drop(
    columns=[
        'user_id',
        'device_id'
    ],
    errors='ignore'
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    stratify=y,
    test_size=0.2,
    random_state=42
)
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(
    random_state=42
)

X_train_res, y_train_res = smote.fit_resample(
    X_train,
    y_train
)
```

```python
print(
    "Before SMOTE:"
)

print(
    y_train.value_counts()
)

print(
    "After SMOTE:"
)

print(
    y_train_res.value_counts()
)
```

The training dataset exhibited a significant class imbalance, with 93,502 legitimate transactions and 9,814 fraudulent transactions. To address this issue, SMOTE (Synthetic Minority Oversampling Technique) was applied exclusively to the training set. After resampling, both classes contained 93,502 observations, creating a balanced dataset for model training while preserving the original distribution of the test set. SMOTE was chosen over undersampling because it retains all majority-class examples and reduces the risk of losing potentially valuable information.
